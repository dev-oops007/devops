```python
#version 2.0 
import requests
import pandas as pd
import ipaddress

#fetching AWS account IPs from CSV
csvData= pd.read_csv('https://yhtpythontestsample.s3.us-east-1.amazonaws.com/IP_List.csv')
myAwsIp= csvData['IPADDRESS'].tolist()
#print('myAWSIP:',myAwsIp)

#newrelic
nrUrl= 'https://s3.amazonaws.com/nr-synthetics-assets/nat-ip-dnsname/production/ip-ranges.json'
#microsoft urls
msUrl = "https://endpoints.office.com/endpoints/worldwide?clientrequestid=b10c5ed1-bad1-445f-b386-b919946339a7"

# fetchig josn data from microsoft URL 
msResponse = requests.get(msUrl)
#print (msResponse)
if msResponse.status_code == 200:
    msIpData= msResponse.json()
msIpList = []

for entry in msIpData:
    if 'ips' in entry:
        msIpList.extend(entry['ips'])
'''   
    if 'urls' in entry:
         print (entry['urls'])    
    else:
         print('No Urls there') 
'''
#for ip in msIpList:
 #   print (ip)

#fetching NewRelic IP
nrResponse = requests.get(nrUrl)
if nrResponse.status_code == 200:
    nrIpData= nrResponse.json()
nrIpList= []
for city, ipList in nrIpData.items():
    nrIpList.extend(ipList)

#for ip in nrIpList:
#    print(ip)

#print(nrIpList)
whitelistIp= []

# Checking AWS IPs against NewRelic & microsoft IP ranges
for aws_ip in myAwsIp:
    aws_ip_obj = ipaddress.ip_address(aws_ip)  # Convert AWS IP to an IP object
    
    # Check if AWS IP is in any Microsoft CIDR range
    for ms_ip in msIpList:
        if aws_ip_obj in ipaddress.ip_network(ms_ip):
            whitelistIp.append(aws_ip)
            break  # Exit loop if match is found

    # Check if AWS IP is in any NewRelic CIDR range
    for nr_ip in nrIpList:
        if aws_ip_obj in ipaddress.ip_network(nr_ip):
            whitelistIp.append(aws_ip)
            break  # Exit loop if match is found

print("Whitelisted IPs:", whitelistIp)

```
